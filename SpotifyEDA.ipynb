```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the dataset using pandas read_csv
dataset = pd.read_csv('tracks.csv')
dataset.head()
```

```python
#Checking number of null values for each column
pd.isnull(dataset).sum()
```

```python
#Checking the info about the dataset.
dataset.info()
```

```python
#Finding the least popular songs in the dataset
sorted_dataset = dataset.sort_values('popularity', ascending = True)
sorted_dataset.head()
```

```python
dataset.describe().transpose()
```

```python
#Finding the most popular songs who have populariy greater than 95
#using pandas query to find populaity greater than 95. Setting inplace as False to not update the values.
mostpopular = dataset.query('popularity>95',inplace=False).sort_values('popularity',ascending=False)
mostpopular[:10]
```

```python
#Changing the index column to release_date and setting the release date to date time format
dataset.set_index('release_date',inplace=True)
dataset.index = pd.to_datetime(dataset.index)
dataset.head()
```

```python
#identifying NaN values using boolean indesing
dataset = dataset[~dataset['duration_ms'].isnull()]
dataset.duration_ms.head()
```

```python
#Converting the duration from ms to seconds
dataset['duration'] = dataset['duration_ms'].apply(lambda x: round (x/1000.0))
dataset.duration.head()
```

```python
#Dropping the duration_ms column
dataset.drop('duration_ms',inplace=True,axis=1)
```

```python
data = dataset.drop(['key','mode','explicit'],axis=1)
```

```python
sample_data = dataset.sample(int(0.004*len(dataset)))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_data,y='loudness',x='energy',color='c').set(title='Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_data,y='popularity',x='acousticness',color='r').set(title='Popularity vs Acouticness Correlation')
```

```python
dataset['dates']=dataset.index.get_level_values('release_date')
dataset.dates=pd.to_datetime(dataset.dates)
years=dataset.dates.dt.year
```

```python
dataset_genre= pd.read_csv('SpotifyFeatures.csv')
dataset_genre.head()
```

```python
plt.title('Duration of the songs in Different Genres')
sns.color_palette('rocket',as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=dataset_genre)
plt.xlabel('Duration in MS')
plt.ylabel('Genres')
```

```python
#Top 5 genres by popularity
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
most= dataset_genre.sort_values('popularity',ascending=False).head(10)
sns.barplot(x='popularity',y='genre',data=most).set(title='Top 5 Genres by popularity')
```
